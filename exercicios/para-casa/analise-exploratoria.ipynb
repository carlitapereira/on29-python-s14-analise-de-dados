```python
import pandas as pd
import os

# Caminhos dos arquivos CSV
customers_file_path = 'datasets/olist_customers_dataset.csv'
orders_file_path = 'datasets/olist_orders_dataset.csv'
output_file_path = 'datasets/olist_customers_orders_merged.csv'

# VerificaÃ§Ã£o da existÃªncia dos arquivos
if not os.path.exists(customers_file_path):
    print(f"Arquivo {customers_file_path} nÃ£o encontrado.")
if not os.path.exists(orders_file_path):
    print(f"Arquivo {orders_file_path} nÃ£o encontrado.")

# Leitura dos arquivos CSV
customers_df = pd.read_csv(customers_file_path)
orders_df = pd.read_csv(orders_file_path)

# UniÃ£o das tabelas utilizando a coluna 'customer_id'
merged_df = pd.merge(customers_df, orders_df, on='customer_id', how='inner')

# Salvando a tabela resultante em um novo arquivo CSV
merged_df.to_csv(output_file_path, index=False)

# ExibiÃ§Ã£o das primeiras linhas da tabela resultante para verificar a uniÃ£o
print(merged_df.head())
```

```python
import pandas as pd
import os

# Caminhos dos arquivos CSV
customers_orders_file_path = 'datasets/olist_customers_orders_merged.csv'
items_file_path = 'datasets/olist_order_items_dataset.csv'

# VerificaÃ§Ã£o da existÃªncia dos arquivos
if not os.path.exists(customers_orders_file_path):
    print(f"Arquivo {customers_orders_file_path} nÃ£o encontrado.")
if not os.path.exists(items_file_path):
    print(f"Arquivo {items_file_path} nÃ£o encontrado.")

# Leitura dos arquivos CSV
customers_orders_df = pd.read_csv(customers_orders_file_path)
items_df = pd.read_csv(items_file_path)

# UniÃ£o das tabelas utilizando a coluna 'order_id'
# Incluindo apenas a coluna 'price' do 'items_df'
final_df = pd.merge(customers_orders_df, items_df[['order_id', 'price']], on='order_id', how='left')

# Caminho para o arquivo final unificado
final_output_file_path = 'datasets/olist_customers_orders_items_merged.csv'

# Salvando a tabela resultante em um novo arquivo CSV
final_df.to_csv(final_output_file_path, index=False)

# ExibiÃ§Ã£o das primeiras linhas da tabela resultante para verificar a uniÃ£o
print(final_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho para o arquivo CSV
file_path = 'datasets/olist_customers_orders_items_merged.csv'

# Carregar a planilha
try:
    df = pd.read_csv(file_path)
    print("Planilha carregada com sucesso.")
except Exception as e:
    print(f"Erro ao carregar a planilha: {e}")

# Verificar se o DataFrame foi carregado corretamente
if 'df' in locals():
    # Agrupar por 'customer_state' e contar o nÃºmero de 'order_id'
    grouped_data = df.groupby('customer_state')['order_id'].count()

    # Criar o grÃ¡fico
    plt.figure(figsize=(12, 6))
    grouped_data.plot(kind='bar', color='skyblue')
    plt.title('NÃºmero de Orders por Estado do Cliente')
    plt.xlabel('Estado do Cliente')
    plt.ylabel('NÃºmero de Orders')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Exibir o grÃ¡fico
    plt.show()
else:
    print("O DataFrame nÃ£o foi carregado.")
```
